```python
import os
import dcex
from dotenv import load_dotenv

load_dotenv()
```

```python
BYBIT_API_KEY = os.getenv("BYBIT_API_KEY")
BYBIT_API_SECRET = os.getenv("BYBIT_API_SECRET")

client = dcex.bybit(
    api_key=BYBIT_API_KEY,
    api_secret=BYBIT_API_SECRET,
)
```

## account http


```python
client.set_margin_mode(margin_mode="ISOLATED_MARGIN")
```

```python
client.repay_liability()
```

## asset http


```python
client.create_internal_transfer("USDT", "1", "FUND", "UNIFIED")
```

```python
client.withdraw(
    coin="USDT",
    chain="APTOS",
    address="",
    amount="1",
)
```

```python
client.cancel_withdrawal("139715944")
```

```python
client.create_universal_transfer(
    coin="USDT",
    amount="4",
    fromMemberId="32881242",
    toMemberId="439606179",
    fromAccountType="FUND",
    toAccountType="FUND",
)
```

```python
client.get_sub_deposit_records("1")
```

```python
client.set_deposit_account(accountType="FUND")
```
